```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
import re
```

```python
url = 'https://www.rottentomatoes.com/m/ballerina_2025/reviews/all-audience'

option = Options()
option.add_argument("start-maximized")

driver = webdriver.Chrome(options=option)
# driver.implicitly_wait(10)
driver.get(url)

```

```python
load_more_1 = driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]')

while True:
    try:
        load_more_1.click()
        # time.sleep(10)
    except Exception as e:
        break
```

```python
load_more_2 = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[2]/rt-button')

while True:
    try:
        load_more_2.click()
        time.sleep(3)
    except Exception as e:
        break
```

```python
# User ID, Review Date, Info

user_ids = []
timestamps = []
reviews = []

for i in range(1, 931):
    user_id = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/rt-link[2]'.format(1*i)).text
    timestamp = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/span[@slot="timestamp"]'.format(1*i)).text
    review = driver.find_element(By.XPATH, '//*[@id="main-page-content"]/div/section/div/div[2]/div[1]/review-card[{}]/drawer-more/span'.format(1*i)).text

    user_ids.append(user_id)
    timestamps.append(timestamp)
    reviews.append(review)
```

```python
print(len(user_ids), len(timestamps), len(reviews))
```

```python
ballerina_raw = {'User_IDs': user_ids, 
                 'Timestamps': timestamps,
                 'Reviews': reviews}

ballerina_raw_df = pd.DataFrame(ballerina_raw)
```

```python
ballerina_raw_df.to_csv("ballerina_raw.csv")
```
